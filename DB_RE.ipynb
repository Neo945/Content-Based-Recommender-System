```python
#import for CountVectorizer and cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Database connectivity
import sqlite3
conn = sqlite3.connect('movies.db')
```

```python
cur = conn.execute("""SELECT "title","genres","cast","director" FROM mytable order by "indexI";""")
```

```python
# Retrieving the data in an array
arr = []
for row in cur:
    stri = ""
    if row[0] != None:
        stri = row[0]
    if row[1] != None:
        stri = stri + " " + row[1]
    if row[2] != None:
        stri = stri + " " + row[2]
    if row[3] != None:
        stri = stri + " " + row[3]
    arr.append(stri)
```

```python
print(arr)
```

```python
cv = CountVectorizer()
countMat = cv.fit_transform(arr)
print(countMat.toarray())
```

```python
# cosine similarity between each string and all the other string in the array
similarity_cos = cosine_similarity(countMat)
print(similarity_cos)
```

```python
# Getting the ID of desired movie
like = """ "Star Wars: Episode III - Revenge of the Sith" """
statement = """select "indexI" from mytable where "title" =  """ + like + """;"""
cur = conn.execute(statement)
```

```python
k = None
for row in cur:
    k = int(row[0])
print(similarity_cos[k])
```

```python
# assigning an index value to each similarity
similar_movies = list(enumerate(similarity_cos[k]))
print(similar_movies)
```

```python
# sorting of list
sorted_similar_mov = sorted(similar_movies,key=lambda x:x[1],reverse=True)
print(sorted_similar_mov)
```

```python
# retriving top 10 results
i = 0
for mov in sorted_similar_mov:
    cur = conn.execute("""select "title" from mytable where "indexI" =  """ + str(mov[0]) + """;""")
    for row in cur:
        print(row[0])
    i = i+1
    if i>10:
        break
```
